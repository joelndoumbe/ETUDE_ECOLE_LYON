# Chargement des données des écoles

```python
import pandas as pd

# Données des écoles
ecoles = pd.read_csv('fr-en-ecoles-effectifs-nb_classes.csv', encoding='utf-8', sep=';')
ecoles.head()
```

```python
# Adresse des écoles
adresses = pd.read_csv('fr-en-annuaire-education.csv', encoding='utf-8', sep=';')
useful_columns = ['Identifiant_de_l_etablissement','Adresse_1', 'Adresse_2', 'Adresse_3', 'Code postal', 'Nom_commune']
adresses = adresses[useful_columns]

adresses.head()
```

```python
# Rattachement de l'adresses aux écoles via leur identifiants
data_ecoles = pd.merge(ecoles, adresses, left_on="Numéro de l'école", right_on="Identifiant_de_l_etablissement")
data_ecoles.columns
```

```python
# On garde uniquement les features qui nous interessent
features_to_keep = ['Rentrée scolaire', 'Dénomination principale', 'Patronyme', 'Secteur', 'Nombre total de classes',
                    "Nombre total d'élèves", "Nombre d'élèves en pré-élémentaire hors ULIS",
                   "Nombre d'élèves en élémentaire hors ULIS", "Nombre d'élèves en ULIS",
                   "Nombre d'élèves en CP hors ULIS", "Nombre d'élèves en CE1 hors ULIS",
                   "Nombre d'élèves en CE2 hors ULIS", "Nombre d'élèves en CM1 hors ULIS",
                   "Nombre d'élèves en CM2 hors ULIS", 'Adresse_1', 'Adresse_2', 'Adresse_3',
                   'Code postal', 'Nom_commune']
data_ecoles = data_ecoles[features_to_keep]
data_ecoles.head()
```

# Scrapping des longitudes et latitudes des écoles

```python
# On crée une colonne avec l'adresse complète
adresses = []
for elt in data_ecoles_metropole[['Adresse_1', 'Adresse_3']].values:
    adresses.append(elt[0] + ' ' + str(elt[1]))

data_ecoles_metropole['adresse_complete'] = adresses
```

```python
#On récupère les adresses
unique_adresses = data_ecoles_metropole['adresse_complete'].unique()
print(len(unique_adresses))
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager

import time

#Chargement de la page web avec selenium
driver = webdriver.Chrome(ChromeDriverManager().install())
driver.maximize_window()
driver.get("https://www.coordonnees-gps.fr/")
driver.refresh()
```

```python
# Form a remplir avec l'adresse
adress_input = '//*[@id="address"]'
find_button = '//*[@id="wrap"]/div[2]/div[2]/div[1]/form[1]/div[2]/div/button'

# Form ou sont stocké lat et long dans la page web
lat_field = '//*[@id="latitude"]'
long_field = '//*[@id="longitude"]'
```

```python
lats = []
longs = []

# Pour chaque adresse dans le dataframe
for i,adresse in enumerate(unique_adresses):
    
    # On rentre l'adresse
    driver.find_element(By.XPATH, adress_input).clear()
    driver.find_element(By.XPATH, adress_input).send_keys(adresse)
    driver.find_element(By.XPATH, find_button).click()
    
    #On attend que la page charge
    time.sleep(1)
    
    #On récupère latitude et longitude
    lat = driver.find_element(By.XPATH, lat_field).get_attribute("value")
    long = driver.find_element(By.XPATH, long_field).get_attribute("value")
    
    #On les ajoute dans leurs listes respectives
    lats.append(lat)
    longs.append(long)
    
```

```python
#Vérification

print(len(lats))
print(len(longs))
print(len(unique_adresses))
```

```python
#On mappe chaque adresse avec ses coordonées dans un dictionnaire
adresse_mapping = {}
for i,elt in enumerate(unique_adresses):
    adresse_mapping[elt] = [lats[i], longs[i]]

# On les ajoutes au dataframe
lats_column = []
longs_column = []
for adresse in data_ecoles_metropole['adresse_complete'].values:
    lats_column.append(adresse_mapping[adresse][0])
    longs_column.append(adresse_mapping[adresse][1])    

data_ecoles_metropole['lat'] = lats_column
data_ecoles_metropole['lon'] = longs_column
```

```python
data_ecoles_metropole.reset_index(inplace=True)
data_ecoles_metropole = data_ecoles_metropole.drop('index', axis=1)
data_ecoles_metropole.head(10)
```

```python
for index in indexes:
    row = data_ecoles_metropole.iloc[index]
    adresse = row['Adresse_1'] + ' ' + str(row['Code postal']) + ' ' + row['Nom_commune']
    
    # On rentre l'adresse
    driver.find_element(By.XPATH, adress_input).clear()
    driver.find_element(By.XPATH, adress_input).send_keys(adresse)
    driver.find_element(By.XPATH, find_button).click()
    
    #On attend que la page charge
    time.sleep(1)
    
    #On récupère latitude et longitude
    lat = driver.find_element(By.XPATH, lat_field).get_attribute("value")
    long = driver.find_element(By.XPATH, long_field).get_attribute("value")
    
    #On modifier avec les nouvelles coordonnées
    data_ecoles_metropole.at[index, 'lat'] = lat
    data_ecoles_metropole.at[index, 'lon'] = long
```

```python
#Vérification
sorted(data_ecoles_metropole['lon'].unique())
```

```python
#Sauvegarde des données
data_ecoles_metropole.to_csv('data_ecoles_metropole.csv', header=True, index=False)
```
